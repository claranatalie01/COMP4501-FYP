## Phase 1 - Data Audit

```python
# df_dompet is already loaded — sanity check
print(f"Shape: {df_dompet.shape[0]:,} rows × {df_dompet.shape[1]} columns")

# ── dtype + null summary ──────────────────────────────────────────
schema = pd.DataFrame({
    'dtype'   : df_dompet.dtypes,
    'null_count': df_dompet.isnull().sum(),
    'null_pct' : (df_dompet.isnull().sum() / len(df_dompet) * 100).round(2),
    'nunique'  : df_dompet.nunique(),
    'sample'   : df_dompet.iloc[0],
})
schema = schema.sort_values('null_pct', ascending=False)

# Flag columns with >5% nulls
schema['flag'] = schema['null_pct'].apply(lambda x: '⚠ HIGH NULL' if x > 5 else ('· low null' if x > 0 else ''))

display(schema.style
    .background_gradient(subset=['null_pct'], cmap='Oranges')
    .set_caption('Schema & Null Audit — df_dompet')
    .format({'null_pct': '{:.2f}%'})
)
```

```python
# ── Duplicates ────────────────────────────────────────────────────
n_dup_rows  = df_dompet.duplicated().sum()
n_dup_ids   = df_dompet['id'].duplicated().sum()
n_dup_title = df_dompet['title'].duplicated().sum()

print(f"Fully duplicate rows : {n_dup_rows}")
print(f"Duplicate product IDs: {n_dup_ids}")
print(f"Duplicate titles     : {n_dup_title}  (may be variant listings)")

# Show any duplicate IDs if they exist
if n_dup_ids > 0:
    dup_sample = df_dompet[df_dompet.duplicated(subset='id', keep=False)]
    display(dup_sample[['id', 'title', 'seller_name', 'final_price']].head(10))
```

```python
# ── Fix known type issues ─────────────────────────────────────────
df = df_dompet.copy()  # work on a copy from here on

# seller_joined_date: likely stored as float (epoch days or timestamp)
if df['seller_joined_date'].dtype != 'datetime64[ns]':
    try:
        df['seller_joined_date'] = pd.to_datetime(df['seller_joined_date'], unit='s')
        print("seller_joined_date → parsed as Unix timestamp (seconds)")
    except:
        df['seller_joined_date'] = pd.to_datetime(df['seller_joined_date'])
        print("seller_joined_date → parsed as datetime string")

# voucher_status: treat as categorical flag (0/1)
df['voucher_status'] = df['voucher_status'].astype('int8')

# video_count: should be int
df['video_count'] = df['video_count'].fillna(0).astype('int8')

# Categorical string cols — strip whitespace & normalise case
cat_cols = ['leather_texture', 'leather_finish', 'leather_type',
            'shipped_from_region', 'wallet_type']
for col in cat_cols:
    df[col] = df[col].str.strip().str.title()

binary_cols = ['coin_slot']
for col in binary_cols:
    df[col] = df[col].map({'Ya': 1, 'Tidak': 0}).astype('int8')

print("\nTypes after corrections:")
print(df[cat_cols + ['seller_joined_date', 'voucher_status', 'video_count', 'coin_slot']].dtypes)
```

```python
# ── Flag business-logic violations ────────────────────────────────
checks = {
    'rating == 0 (inactive/new)'        : (df['rating'] == 0).sum(),
    'reviews == 0'                      : (df['reviews'] == 0).sum(),
    'sold == 0'                         : (df['sold'] == 0).sum(),
    'stock == 0 (out-of-stock)'         : (df['stock'] == 0).sum(),
    'final_price > initial_price'       : (df['final_price'] > df['initial_price']).sum(),
    'discount == 0 but price changed'   : ((df['discount'] == 0) &
                                             (df['final_price'] != df['initial_price'])).sum(),
    'gmv_cal == 0'                      : (df['gmv_cal'] == 0).sum(),
    'image_count < 3 (below Shopee min)': (df['image_count'] < 3).sum(),
    'seller_rating < 4.0'               : (df['seller_rating'] < 4.0).sum(),
}

check_df = pd.DataFrame(checks.items(), columns=['Check', 'Count'])
check_df['Pct of dataset'] = (check_df['Count'] / len(df) * 100).round(2)
check_df['Action'] = check_df['Count'].apply(
    lambda x: '✓ None found' if x == 0 else '⚠ Review'
)
display(check_df.style.set_caption('Business Logic Violation Audit'))
```

```python
# ── IQR-based outlier scan on key numeric cols ────────────────────
numeric_cols = ['final_price', 'initial_price', 'reviews', 'sold',
                'stock', 'favorite', 'gmv_cal', 'seller_followers', 'rating']

outlier_report = []
for col in numeric_cols:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lo, hi = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    n_out = ((df[col] < lo) | (df[col] > hi)).sum()
    outlier_report.append({
        'column'      : col,
        'Q1'          : round(Q1, 1),
        'Q3'          : round(Q3, 1),
        'IQR_lower'   : round(lo, 1),
        'IQR_upper'   : round(hi, 1),
        'max_value'   : round(df[col].max(), 1),
        'n_outliers'  : n_out,
        'outlier_pct' : round(n_out / len(df) * 100, 2),
    })

out_df = pd.DataFrame(outlier_report)
display(out_df.style
    .background_gradient(subset=['outlier_pct'], cmap='Reds')
    .set_caption('IQR Outlier Scan — Key Numeric Columns')
    .format({'outlier_pct': '{:.2f}%'})
)

# Note: do NOT remove outliers yet — top sellers ARE outliers by design.
# Log-transformation (Phase 6) handles this more appropriately.
```

### Further Clean Data Based on Audit

```python
audit_df = df.copy()

# Step 1: drop fake listings (0 sold, but has reviews — likely relisted)
audit_df = audit_df[~((audit_df['sold'] == 0) & (audit_df['reviews'] > 0))]

# Step 2: deduplicate by ID, keep highest sold
audit_df = (audit_df
            .sort_values('sold', ascending=False)
            .drop_duplicates(subset='id', keep='first')
            .copy())

# Step 3: drop price outliers
audit_df = audit_df[audit_df['final_price'] <= 975_000].copy()

# Step 4: manipulate listings that have sold == 0 to be 

print(f"Final clean dataset: {len(audit_df):,} rows")
```

```python
mask = (audit_df['reviews'] > audit_df['sold']) & (audit_df['reviews'] > 50)

audit_df.loc[mask, 'sold'] = (audit_df.loc[mask, 'reviews'] * 1.1).round().astype(int)

print(f"Listings reassigned: {mask.sum():,}")
print(f"\nSample check:")
print(audit_df.loc[mask, ['id', 'title', 'sold', 'reviews']].head(10))
```

```python
# Recalculate GMV for reassigned rows using final_price * new sold
audit_df.loc[mask, 'gmv_cal'] = audit_df.loc[mask, 'final_price'] * audit_df.loc[mask, 'sold']

print(f"\nGMV recalculated for {mask.sum():,} relisted listings")
print(audit_df.loc[mask, ['id', 'sold', 'reviews', 'final_price', 'gmv_cal']].head(5))
```

```python
# ── Master heatmap — which features correlate with GMV? ──────────
corr_cols = [
    'gmv_cal', 'sold', 'reviews', 'favorite', 'rating',
    'final_price', 'discount', 'image_count', 'video_count',
    'variations_count', 'stock', 'voucher_status',
    'seller_rating', 'seller_followers', 'seller_products',
    'seller_chats_responded_percentage', 'seller_chat_time_reply', 
    'coin_slot', 'rating_credibility', 
]

spearman_mx = audit_df[corr_cols].corr(method='spearman')

fig, ax = plt.subplots(figsize=(14, 11))
mask = np.triu(np.ones_like(spearman_mx, dtype=bool))
sns.heatmap(
    spearman_mx, mask=mask,
    annot=True, fmt='.2f', annot_kws={'size': 8},
    cmap='RdYlGn', center=0, vmin=-1, vmax=1,
    linewidths=0.4, linecolor='white', square=True, ax=ax
)
ax.set_title('Spearman correlation matrix — cleaned wallet listings', fontsize=12, pad=12)
ax.tick_params(axis='x', rotation=45)
ax.tick_params(axis='y', rotation=0)
plt.tight_layout()
plt.show()

sold_corr = spearman_mx['sold'].drop('sold').sort_values(ascending=False)
print("\n── Spearman ρ vs sold (ranked) ──")
for feat, val in sold_corr.items():
    bar  = '█' * int(abs(val) * 20)
    sign = '+' if val >= 0 else ''
    print(f"  {feat:40} {sign}{val:.3f}  {bar}")

rating_corr = spearman_mx['rating'].drop('rating').sort_values(ascending=False)
print("\n── Spearman ρ vs rating (ranked) ──")
for feat, val in rating_corr.items():
    bar  = '█' * int(abs(val) * 20)
    sign = '+' if val >= 0 else ''
    print(f"  {feat:40} {sign}{val:.3f}  {bar}")


rating_cred_corr = spearman_mx['rating_credibility'].drop('rating_credibility').sort_values(ascending=False)
print("\n── Spearman ρ vs rating_credibility (ranked) ──")
for feat, val in rating_cred_corr.items():
    bar  = '█' * int(abs(val) * 20)
    sign = '+' if val >= 0 else ''
    print(f"  {feat:40} {sign}{val:.3f}  {bar}")
```
